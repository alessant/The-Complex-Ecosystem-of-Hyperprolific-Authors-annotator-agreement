```python
import pandas as pd
import numpy as np
```

```python
fpath = "../data/related_behaviors_works/included_works_dealing_w_related_behaviors.csv"
```

```python
df = pd.read_csv(fpath)
```

```python
df.head()
```

```python
# count how many papers there are per year
papers_per_year = df.groupby('Year').size().reset_index(name='n-papers')
```

```python
papers_per_year
```

```python
# fill the gap for the years that have no papers
years = np.arange(df['Year'].min(), df['Year'].max() + 1)
papers_per_year = papers_per_year.set_index('Year').reindex(years, fill_value=0).reset_index()
```

```python
papers_per_year
```

```python
# save df
papers_per_year.to_csv('papers_per_year.csv', index=True)
```
