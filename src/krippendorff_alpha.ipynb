```python
import pandas as pd
import numpy as np
import krippendorff
```

```python
def get_reliability_data(filepath, columns):
    df = pd.read_csv(filepath, usecols=columns)
    df.fillna(np.nan, inplace=True)
    data = df.to_numpy().T.tolist()
    # convert each entry to int if not nan
    data = [[int(entry) if not pd.isna(entry) else np.nan for entry in row] for row in data]
    return data
    
```

### Step 2: elegibility phase

```python
columns_of_interest_step2 = ["Reviewer1", "Reviewer2", "Reviewer3", "Reviewer4", "Reviewer5"]
scoref_step2 = "../data/step2_eligibility.csv"

reliability_data_step2 = get_reliability_data(scoref_step2, columns_of_interest_step2)

print(
    "Krippendorff's alpha for nominal metric: ",
    krippendorff.alpha(reliability_data=reliability_data_step2, level_of_measurement="nominal"),
)
```
