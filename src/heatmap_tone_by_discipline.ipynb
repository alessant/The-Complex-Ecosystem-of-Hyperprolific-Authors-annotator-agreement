```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Data - see Table 2 from the article
data = pd.DataFrame(
    [
        [33, 0, 0, 40, 22],
        [33, 0, 60, 0, 28],
        [17, 100, 40, 40, 39],
        [17, 0, 0, 20, 11],
    ],
    index=["Accusatory", "Ambiguous", "Neutral", "Optimistic"],
    columns=["Medical\nsciences", "Computer\nscience", "Across\ndomains", "Multi-\ndomain", "Overall"]
)
```

```python
data
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "axes.labelsize": 14,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
})
```

```python
# Plot
plt.figure(figsize=(5.7, 3.5))
ax = sns.heatmap(
    data,
    cmap="Blues",
    annot=True,
    fmt="d",
    linewidths=0.5,
    annot_kws={"size": 14},
    cbar=False,
    cbar_kws={"label": "Percentage"}
)

# do not show ticks
ax.tick_params(left=False, bottom=False)

# Add percent sign to annotations
for text in ax.texts:
    text.set_text(text.get_text() + "\\%")
    # value = int(text.get_text().replace("\%", ""))

ax.set_xticklabels(
    ax.get_xticklabels(),
    fontsize=14
)  

ax.set_yticklabels(
    ax.get_yticklabels(),   
    fontsize=14,
)

# Labels
# ax.set_xlabel("Discipline")
# ax.set_ylabel("Tone")
plt.tight_layout()

# save the figure 
plt.savefig("heatmap_tone_discipline.pdf", bbox_inches="tight")

# save the figure in eps format
plt.savefig("heatmap_tone_discipline.eps", format="eps", bbox_inches="tight")
plt.show()
```
